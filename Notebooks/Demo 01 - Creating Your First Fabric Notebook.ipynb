##### Fabric support for Python includes mssparkutils module
 - **mssparkutils** provides library of convenient utility APIs to increase productivity
 - Direct access to file system of Lakehouse available using **mssparkutils.fs** object
 - You can use **mssparkutils.fs.put** to create new file in Lakehouse storage
 - You must first associate this notebook with a Fabric Lakehouse for this to work  


```python
message = "And now for something completely different"
file_path = "Files/demo01/hello.txt"
mssparkutils.fs.put(file_path, message, True)
```
